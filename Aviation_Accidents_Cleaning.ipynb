## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:
- inspect relevant columns
- figure out any reasonable imputations
- filter the dataset

```python
# Import libraries again
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data again
df = pd.read_csv('data/AviationData.csv', encoding='latin1', low_memory=False)

print("Libraries imported and data loaded successfully!")
print(f"Dataset shape: {df.shape}")
```

```python
# Step 3: Filter dataset to include aircraft from 1983 onwards
# Client assumes max lifetime of 40 years for make/model retirement (2023 - 40 = 1983)

print("=" * 80)
print("FILTERING BY YEAR (1983 onwards)")
print("=" * 80)

# Convert Event.Date to datetime and extract year
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df['Year'] = df['Event.Date'].dt.year

# Check year range
print(f"Year range: {df['Year'].min()} to {df['Year'].max()}")
print(f"Total records: {len(df)}")

# Filter for years 1983 onwards
df_filtered = df[df['Year'] >= 1983].copy()

print(f"\nFiltered dataset shape (1983+): {df_filtered.shape}")
print(f"Rows removed: {len(df) - len(df_filtered)}")
print(f"Percentage of data retained: {(len(df_filtered)/len(df))*100:.2f}%")

# Preview the filtered data
print("\nFirst few rows of filtered data:")
display(df_filtered.head(3))
```

### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
# Step 4: Clean injury columns and create derived metrics

print("=" * 80)
print("CLEANING INJURY COLUMNS")
print("=" * 80)

# Convert injury columns to numeric, filling NaN with 0
injury_cols = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']

for col in injury_cols:
    df_filtered[col] = pd.to_numeric(df_filtered[col], errors='coerce').fillna(0)
    print(f"Cleaned: {col}")

# Calculate total people aboard
df_filtered['Total_Aboard'] = df_filtered[injury_cols].sum(axis=1)

print(f"\nTotal_Aboard column created")
print(f"Summary statistics for Total_Aboard:")
print(df_filtered['Total_Aboard'].describe())

# Create metric: Serious/Fatal injury count
df_filtered['Serious_Fatal_Injuries'] = df_filtered['Total.Fatal.Injuries'] + df_filtered['Total.Serious.Injuries']
print(f"\nSerious_Fatal_Injuries column created")

# Calculate injury rate (percentage of people seriously injured or killed)
df_filtered['Injury_Rate'] = np.where(
    df_filtered['Total_Aboard'] > 0,
    (df_filtered['Serious_Fatal_Injuries'] / df_filtered['Total_Aboard']) * 100,
    0
)

print(f"Injury_Rate column created (percentage)")

print("\nInjury Rate Summary Statistics (%):")
print(df_filtered['Injury_Rate'].describe())

# Check for any extreme values
print(f"\nRecords with >100% injury rate: {(df_filtered['Injury_Rate'] > 100).sum()}")
```

**Aircraft.Damage**
- identify and execute any cleaning tasks
- construct a derived column tracking whether an aircraft was destroyed or not.

```python
# Step 5: Clean Aircraft.damage column and create destroyed flag

print("=" * 80)
print("CLEANING AIRCRAFT DAMAGE COLUMN")
print("=" * 80)

# Inspect unique values before cleaning
print("Unique values before cleaning:")
print(df_filtered['Aircraft.damage'].value_counts())

# Clean and standardize damage values
df_filtered['Damage_Cleaned'] = df_filtered['Aircraft.damage'].str.strip().str.lower() if df_filtered['Aircraft.damage'].dtype == 'object' else df_filtered['Aircraft.damage']

# Create destroyed flag
# Keywords indicating aircraft was destroyed or substantially damaged
destroyed_keywords = ['destroyed', 'substantial', 'written off', 'demolished']

df_filtered['Is_Destroyed'] = df_filtered['Damage_Cleaned'].apply(
    lambda x: any(keyword in str(x).lower() for keyword in destroyed_keywords) if pd.notna(x) else False
)

print(f"\nDestroyed flag created:")
print(f"Destroyed: {df_filtered['Is_Destroyed'].sum()}")
print(f"Not Destroyed: {(~df_filtered['Is_Destroyed']).sum()}")
print(f"Unknown: {df_filtered['Damage_Cleaned'].isna().sum()}")
print(f"Percentage destroyed: {(df_filtered['Is_Destroyed'].sum() / len(df_filtered)) * 100:.2f}%")
```

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
# Step 6: Investigate the Make column

print("=" * 80)
print("INVESTIGATING THE MAKE COLUMN")
print("=" * 80)

# Check for missing values
print(f"Missing values in Make column: {df_filtered['Make'].isna().sum()}")

# Inspect unique values and counts
print(f"\nNumber of unique Makes: {df_filtered['Make'].nunique()}")

# Get top 30 Makes by frequency
print("\nTop 30 Most Common Makes:")
make_counts = df_filtered['Make'].value_counts()
print(make_counts.head(30))

# Check for any data quality issues (whitespace, case inconsistencies)
print("\nSample of unique Make values (first 20):")
print(df_filtered['Make'].dropna().unique()[:20])

# Clean the Make column - strip whitespace and standardize
df_filtered['Make_Cleaned'] = df_filtered['Make'].str.strip().str.upper() if df_filtered['Make'].dtype == 'object' else df_filtered['Make']

# Check for makes that might be the same but spelled differently
print("\nChecking for potential duplicate makes (close matches):")
# Look for common makes that might appear with variations
common_makes = ['BOEING', 'AIRBUS', 'CESSNA', 'PIPER', 'BEECHCRAFT', 'MOONEY', 'BELL', 'AIR TRACTOR']
for make in common_makes:
    matches = df_filtered[df_filtered['Make_Cleaned'].str.contains(make, na=False)]['Make_Cleaned'].unique()
    if len(matches) > 1:
        print(f"{make}: {matches}")
    elif len(matches) == 1:
        print(f"{make}: {matches[0]}")

# Filter to keep Makes with reasonable number (threshold >= 50)
make_threshold = 50
valid_makes = make_counts[make_counts >= make_threshold].index.tolist()
print(f"\nMakes with >= {make_threshold} occurrences: {len(valid_makes)}")
print(f"First 20: {valid_makes[:20]}")

# Create filtered dataframe for analysis
df_analysis = df_filtered[df_filtered['Make'].isin(valid_makes)].copy()
print(f"\nDataframe shape after filtering to valid makes: {df_analysis.shape}")
print(f"Original shape: {df_filtered.shape}")
print(f"Rows removed: {len(df_filtered) - len(df_analysis)}")
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
# Step 7: Inspect Model column

print("=" * 80)
print("INSPECTING THE MODEL COLUMN")
print("=" * 80)

# Check for missing values
print(f"Missing values in Model column: {df_analysis['Model'].isna().sum()}")

# Drop rows with missing Model
df_analysis = df_analysis.dropna(subset=['Model']).copy()
print(f"Shape after dropping missing models: {df_analysis.shape}")

# Inspect model counts
print(f"\nNumber of unique Models: {df_analysis['Model'].nunique()}")

# Check if model labels are unique to each make
print("\nChecking if Model names are unique to each Make:")
# Look for models that appear under different makes
model_make_groups = df_analysis.groupby('Model')['Make_Cleaned'].nunique()
models_multiple_makes = model_make_groups[model_make_groups > 1]
print(f"Models that appear under multiple Makes: {len(models_multiple_makes)}")

if len(models_multiple_makes) > 0:
    print("\nExamples of models with multiple makes:")
    for model in models_multiple_makes.head(10).index:
        makes = df_analysis[df_analysis['Model'] == model]['Make_Cleaned'].unique()
        print(f"  '{model}': {makes}")

# Create a unique identifier for plane type (Make + Model)
df_analysis['Plane_Type'] = df_analysis['Make_Cleaned'] + ' - ' + df_analysis['Model']
print(f"\nCreated 'Plane_Type' column (Make + Model combination)")
print(f"Number of unique plane types: {df_analysis['Plane_Type'].nunique()}")

# Get top plane types by frequency
print("\nTop 20 Most Common Plane Types:")
plane_type_counts = df_analysis['Plane_Type'].value_counts()
print(plane_type_counts.head(20))
```

```python
# Step 8: Cleaning other columns (Engine.Type, Weather.Condition, Number.of.Engines, Purpose.of.flight, Broad.phase.of.flight)

print("=" * 80)
print("CLEANING OTHER COLUMNS")
print("=" * 80)

# 8.1 Engine.Type
print("\n" + "-" * 40)
print("ENGINE.TYPE COLUMN")
print("-" * 40)
print(f"Unique values before cleaning:")
print(df_analysis['Engine.Type'].value_counts())
print(f"Missing values: {df_analysis['Engine.Type'].isna().sum()}")

# Clean Engine.Type
df_analysis['Engine_Type_Cleaned'] = df_analysis['Engine.Type'].str.strip().str.upper() if df_analysis['Engine.Type'].dtype == 'object' else df_analysis['Engine.Type']

# 8.2 Weather.Condition
print("\n" + "-" * 40)
print("WEATHER.CONDITION COLUMN")
print("-" * 40)
print(f"Unique values before cleaning:")
print(df_analysis['Weather.Condition'].value_counts())
print(f"Missing values: {df_analysis['Weather.Condition'].isna().sum()}")

# Clean Weather.Condition
df_analysis['Weather_Cleaned'] = df_analysis['Weather.Condition'].str.strip().str.upper() if df_analysis['Weather.Condition'].dtype == 'object' else df_analysis['Weather.Condition']

# 8.3 Number.of.Engines
print("\n" + "-" * 40)
print("NUMBER.OF.ENGINES COLUMN")
print("-" * 40)
print(f"Summary statistics:")
print(df_analysis['Number.of.Engines'].describe())
print(f"Missing values: {df_analysis['Number.of.Engines'].isna().sum()}")

# Fill missing Number.of.Engines with median (1)
df_analysis['Number_of_Engines_Cleaned'] = df_analysis['Number.of.Engines'].fillna(1)

# 8.4 Purpose.of.flight
print("\n" + "-" * 40)
print("PURPOSE.OF.FLIGHT COLUMN")
print("-" * 40)
print(f"Unique values before cleaning:")
print(df_analysis['Purpose.of.flight'].value_counts())
print(f"Missing values: {df_analysis['Purpose.of.flight'].isna().sum()}")

# Clean Purpose.of.flight
df_analysis['Purpose_Cleaned'] = df_analysis['Purpose.of.flight'].str.strip().str.upper() if df_analysis['Purpose.of.flight'].dtype == 'object' else df_analysis['Purpose.of.flight']

# 8.5 Broad.phase.of.flight
print("\n" + "-" * 40)
print("BROAD.PHASE.OF.FLIGHT COLUMN")
print("-" * 40)
print(f"Unique values before cleaning:")
print(df_analysis['Broad.phase.of.flight'].value_counts())
print(f"Missing values: {df_analysis['Broad.phase.of.flight'].isna().sum()}")

# Clean Broad.phase.of.flight
df_analysis['Phase_Cleaned'] = df_analysis['Broad.phase.of.flight'].str.strip().str.upper() if df_analysis['Broad.phase.of.flight'].dtype == 'object' else df_analysis['Broad.phase.of.flight']

print("\n" + "=" * 80)
print("ALL COLUMNS CLEANED SUCCESSFULLY")
print("=" * 80)
```
